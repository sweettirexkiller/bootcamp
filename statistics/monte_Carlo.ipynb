```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 100
plt.axes()

squere = plt.Rectangle((-1,-1),width=2,height=2, alpha=0.3)
circle = plt.Circle((0,0), radius=1, fc='y', alpha=0.2)

plt.gca().add_patch(squere)
plt.gca().add_patch(circle)

X = np.random.uniform(-1,1, 2*N).reshape(N,2)

inside = X[ (X[:,0]**2 + X[:,1]**2) <= 1]
outside = X[ (X[:,0]**2 + X[:,1]**2) > 1]


plt.scatter(inside[:,0], inside[:,1], color='red', alpha=1)
plt.scatter(outside[:,0], outside[:,1], color='black', alpha=1)


plt.axis('scaled')
plt.show()
```
