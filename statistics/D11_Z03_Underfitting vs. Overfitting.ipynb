# Underfitting vs. Overfitting

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
from sklearn import  metrics
```

Rozważmy następujący zbiór punktów:

```python
np.random.seed(0)
n_samples = 30
true_fun = lambda X: np.cos(1.5 * np.pi * X)
X = np.sort(np.random.rand(n_samples))
y = true_fun(X) + np.random.randn(n_samples) * 0.1
```

```python
# Plot outputs
plt.figure(figsize=(6,6));
axes = plt.gca()
axes.set_xlim([0,1])
axes.set_ylim([-1.5,1.5])
plt.scatter(X, y,  color='black')
x_tr = np.linspace(0, 1, 200)
plt.show()
```

# Zadanie
Podziel dane na dwie grupy.

```python
s=np.random.random_sample(n_samples)
s[s>0.2]=1
s[s<=0.2]=0
s
X1=X[s==1]
y1=y[s==1]
X2=X[s==0]
y2=y[s==0]

# Plot outputs
plt.figure(figsize=(6,6));
axes = plt.gca()
axes.set_xlim([0,1])
axes.set_ylim([-1.5,1.5])
plt.scatter(X1, y1,  color='blue')
plt.scatter(X2, y2,  color='red')
x_tr = np.linspace(0, 1, 200)
plt.show()
```

# Zadanie
Wykonaj regresję:
* liniową


```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
from sklearn import datasets, linear_model
```

```python
model1 = linear_model.LinearRegression()
model1.fit(np.vstack(X1), y1)

model2 = linear_model.LinearRegression()
model2.fit(np.vstack(X2), y2)
```

```python
plt.figure(figsize=(6,6));
axes = plt.gca()
axes.set_xlim([0,1])
axes.set_ylim([-1.5,1.5])
plt.scatter(X1, y1,  color='blue')
plt.scatter(X2, y2,  color='red')

plt.plot(X1, model1.predict(np.vstack(X1)), color='blue',linewidth=3)
plt.plot(X2, model2.predict(np.vstack(X2)), color='red',linewidth=3)


x_tr = np.linspace(0, 1, 200)
plt.show()
```

# Zadanie
Wykonaj regresję:
* wielomianową stopnia $4$


```python
model3 = make_pipeline(PolynomialFeatures(2), linear_model.LinearRegression())
model3.fit(np.vstack(X1), y1)

model4 = make_pipeline(PolynomialFeatures(2), linear_model.LinearRegression())
model4.fit(np.vstack(X2), y2)


plt.figure(figsize=(6,6));
axes = plt.gca()
axes.set_xlim([0,1])
axes.set_ylim([-1.5,1.5])
plt.scatter(X1, y1,  color='blue')
plt.scatter(X2, y2,  color='red')

plt.plot(X1, model3.predict(np.vstack(X1)), color='blue',linewidth=3)
plt.plot(X2, model4.predict(np.vstack(X2)), color='red',linewidth=3)


x_tr = np.linspace(0, 1, 200)
plt.show()
```

# Zadanie
Wykonaj regresję:

```python
model5 = make_pipeline(PolynomialFeatures(15), linear_model.LinearRegression())
model5.fit(np.vstack(X1), y1)

model6 = make_pipeline(PolynomialFeatures(15), linear_model.LinearRegression())
model6.fit(np.vstack(X2), y2)


plt.figure(figsize=(6,6));
axes = plt.gca()
axes.set_xlim([0,1])
axes.set_ylim([-1.5,1.5])
plt.scatter(X1, y1,  color='blue')
plt.scatter(X2, y2,  color='red')

plt.plot(X1, model5.predict(np.vstack(X1)), color='blue',linewidth=3)
plt.plot(X2, model6.predict(np.vstack(X2)), color='red',linewidth=3)


x_tr = np.linspace(0, 1, 200)
plt.show()
```
