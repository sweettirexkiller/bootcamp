My narazie urzyjemy metody opartej na rozwiązaniu układu równań liniowych.

```python
f_r_linear = lambda x: p1[0][1] * x +p1[0][0]
```

```python
x_prediction_linear = np.linspace(0., 3, 200)
y_prediction_linear = f_r_linear(x_tr)
plt.figure(figsize=(6,6));
axes = plt.gca()
axes.set_xlim([0,3])
axes.set_ylim([0,8])
plt.plot(x_prediction_linear, y_prediction_linear, 'g');
plt.plot(x_population, y_population, 'ok', ms=10);
plt.show()
```

# Zadanie
Nauczmy prosty model regresji

$$
y=a_1\cdot x + a_2 \cdot x^2 + b
$$

na naszych danych.

Oraz narysujmy prostą regresji.

```python
M2 = np.vstack( (np.ones_like(x_population), x_population, x_population**2) ).T
p2 = np.linalg.lstsq(M2, y_population)
```

Definiujemy naszą funkcję regresji

```python
f_r_quadratic = lambda x: p2[0][2]*pow(x,2) + p2[0][1] * x + p2[0][0] 
```

```python
x_prediction_quadratic = np.linspace(0., 3, 200)
y_prediction_quadratic = f_r_quadratic(x_prediction_quadratic)
plt.figure(figsize=(6,6));
axes = plt.gca()
axes.set_xlim([0,3])
axes.set_ylim([0,8])
plt.plot(x_prediction_quadratic, y_prediction_quadratic, 'g');
plt.plot(x_population, y_population, 'ok', ms=10);
plt.show()
```
