# Random Forest Regressor

```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import sklearn.linear_model as lm
from sklearn import datasets, linear_model
import matplotlib.pyplot as plt

from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
```

Proszę narysować wykres dowolnej funkcji na odcinku $[0,3]$.

Będzie to model, który będziemy chcieli znaleźć.

```python
f = lambda x: ((x*2-1)*(x**2-2)*(x-2)+3)
x_tr = np.linspace(0, 3, 200)
y_tr = f(x_tr)
x = stats.uniform(0,3).rvs(100)
y = f(x) + stats.norm(0,0.2).rvs(len(x))
```

# Zadanie
wykonaj regresję z różnymi parametrami

```python
x=np.vstack(x)
x_plot = np.vstack(np.linspace(0, 10, 100))
regr_rf_1 = RandomForestRegressor(max_depth=5, random_state=5)
y_rbf_1 = regr_rf_1.fit(x, y)
# Plot outputs
plt.figure(figsize=(6,6));
axes = plt.gca()
axes.set_xlim([0,3])
axes.set_ylim([0,8])
plt.scatter(x, y,  color='black')
plt.plot(x_plot, regr_rf_1.predict(x_plot), color='blue',linewidth=3)
plt.show()
```

```python
x=np.vstack(x)
x_plot = np.vstack(np.linspace(0, 10, 100))
regr_rf_2 = RandomForestRegressor(max_depth=10, random_state=3)
y_rbf_2 = regr_rf_2.fit(x, y)
# Plot outputs
plt.figure(figsize=(6,6));
axes = plt.gca()
axes.set_xlim([0,3])
axes.set_ylim([0,8])
plt.scatter(x, y,  color='black')
plt.plot(x_plot, regr_rf_2.predict(x_plot), color='blue',linewidth=3)
plt.show()
```
